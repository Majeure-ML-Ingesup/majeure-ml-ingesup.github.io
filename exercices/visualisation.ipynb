## Intéressons nous à la variable SalePrice

```python
df['SalePrice'].describe()
```

```python
sns.distplot(df['SalePrice']);
```

```python


#saleprice correlation matrix
corrmat = df.corr() # Méthode Pearson 
```

```python
corrmat
```

```python

k = 10 #nombre de variables à afficher dans le heatmap
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index #on récupère les 10 variables les plus corrélées à SalePrice
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)

fig, ax = plt.subplots(figsize=(10,10))  
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.1f', annot_kws={'size': 8}, yticklabels=cols.values, xticklabels=cols.values, ax=ax)
plt.show()

```

```python
# On va afficher plus d'informations entre la vaiable OverallQual et SalePrice
sns.set()
sns.pairplot(df[['SalePrice', 'OverallQual']], size = 4)
plt.show();
```
