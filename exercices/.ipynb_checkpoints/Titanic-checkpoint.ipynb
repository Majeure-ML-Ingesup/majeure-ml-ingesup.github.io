# Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Data

```python
df = pd.read_csv("Titanic/train.csv")
```

```python
df.head()
```

# Data processing

```python
df.isna().sum().apply(lambda col: str(round(col/len(df), 2)*100) + " %")
```

## ID

```python
df_processed = df.drop(columns=['PassengerId'])
```

## Sex

```python
df_processed['Sex'] = df_processed['Sex'].apply(lambda val: 1 if val == 'male' else 0)
```

## One Hot Encoding (OHE)

```python
ohe_cols_sib = pd.get_dummies(df_processed['SibSp'], dummy_na=True, prefix='SibSp')
ohe_cols_parch = pd.get_dummies(df_processed['Parch'], dummy_na=True, prefix='SibSp')
ohe_cols_embarked = pd.get_dummies(df_processed['Embarked'], dummy_na=True, prefix='SibSp')
```

```python
ohe_df = pd.concat([
    df_processed.drop(columns=['SibSp', 'Parch', 'Embarked']),
    ohe_cols_sib,
    ohe_cols_parch,
    ohe_cols_embarked
], axis=1)
```

```python
n_unique_cols = ohe_df.nunique()
```

```python
one_unique_value_cols = n_unique_cols[n_unique_cols == 1].index
```

```python
df_unique_processed = ohe_df.drop(columns=one_unique_value_cols.values)
```

## Fare

```python
sns.distplot(df_unique_processed['Fare'])
```

```python
sns.pairplot(df_unique_processed[['Fare', 'Survived']], size = 4)
```

```python
df_fare_cut = pd.cut(df_unique_processed['Fare'], 3)
```

```python
df_fare_cut_dummies = pd.get_dummies(df_fare_cut, prefix='Fare')
```

```python
df_fare_processed = pd.concat([
    df_unique_processed.drop(columns=['Fare']),
    df_fare_cut_dummies
], axis=1)
```
